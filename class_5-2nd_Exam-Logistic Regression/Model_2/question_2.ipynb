# Diferenças entre Regressão Linear e Regressão Logística

---

## 1. **Natureza do Problema**
### Regressão Linear:
- Utilizada para problemas de **regressão**, onde o objetivo é prever valores contínuos.
- **Exemplo**: Previsão de preços de casas, temperatura, ou vendas.

### Regressão Logística:
- Utilizada para problemas de **classificação**, onde o objetivo é prever categorias ou classes.
- **Exemplo**: Determinar se um e-mail é spam (0 ou 1), ou prever a aprovação de um empréstimo (sim ou não).

---

## 2. **Variável Dependente**
### Regressão Linear:
- A variável dependente (ou alvo) é **contínua**.
- **Exemplo**: $Y = 150.5$, $Y = 200.3$.

### Regressão Logística:
- A variável dependente é **categórica** ou **binária** (0 ou 1).
- **Exemplo**: $Y = 0$ (não aprovado), $Y = 1$ (aprovado).

---

## 3. **Função Matemática**
### Regressão Linear:
- Baseia-se em uma **relação linear** entre as variáveis independentes e a variável dependente.
- **Fórmula**:  
  $$
  Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \epsilon
  $$
  - Onde:
    - $Y$: Variável dependente.
    - $\beta_0$: Intercepto.
    - $\beta_1, \beta_2$: Coeficientes das variáveis independentes.
    - $\epsilon$: Erro.

### Regressão Logística:
- Baseia-se na **função logística (sigmoide)** para modelar a probabilidade de uma classe.
- **Fórmula**:  
  $$
  P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + \beta_2X_2 + \dots)}}
  $$
  - O resultado é uma probabilidade entre 0 e 1.

---

## 4. **Distribuição dos Resíduos**
### Regressão Linear:
- Assume que os resíduos seguem uma **distribuição normal**.

### Regressão Logística:
- Não assume normalidade dos resíduos, pois trabalha com probabilidades.

---

## 5. **Métrica de Avaliação**
### Regressão Linear:
- Métricas comuns incluem:
  - **Erro Quadrático Médio (MSE)**.
  - **Coeficiente de Determinação ($R^2$)**.

### Regressão Logística:
- Métricas comuns incluem:
  - **Acurácia**.
  - **Precisão, Recall e F1-Score**.
  - **Curva ROC e AUC**.

---

## 6. **Extrapolação**
### Regressão Linear:
- Pode prever valores fora do intervalo dos dados de treino.

### Regressão Logística:
- Restringe as previsões ao intervalo de 0 a 1, representando probabilidades.

---

## 7. **Aplicações**
### Regressão Linear:
- Previsão de valores contínuos, como:
  - Preço de ações.
  - Crescimento populacional.

### Regressão Logística:
- Classificação binária ou multiclasse, como:
  - Diagnóstico médico (doente ou saudável).
  - Classificação de imagens.

---

## Resumo em Tabela

| **Característica**         | **Regressão Linear**     | **Regressão Logística**   |
|----------------------------|--------------------------|---------------------------|
| **Tipo de Problema**       | Regressão (Contínuo)     | Classificação (Binário)   |
| **Variável Dependente**    | Contínua                | Categórica (0 ou 1)       |
| **Função Matemática**      | Linear                  | Logística (Sigmoide)      |
| **Resíduos**               | Distribuição Normal      | Não assume normalidade    |
| **Métricas de Avaliação**  | MSE, $R^2$              | Acurácia, ROC, AUC        |
| **Intervalo de Previsão**  | Sem restrição           | Entre 0 e 1               |

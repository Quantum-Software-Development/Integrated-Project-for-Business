dados do exercício anterior


```python
# Importar bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import statsmodels.api as sm

```

```python
# Criar o dataframe com os dados fornecidos
data = {
    'Peso': [14, 14, 16, 17, 20, 22, 22, 23, 28, 28],
    'Comprimento': [85, 90, 95, 100, 95, 96, 100, 109, 105, 110]
}
df = pd.DataFrame(data)

# Exibir as primeiras linhas do dataframe
df.head()

# Calcular o coeficiente de correlação de Pearson
correlation, p_value = pearsonr(df['Peso'], df['Comprimento'])
print(f"Coeficiente de Correlação de Pearson: {correlation}")
print(f"Valor p: {p_value}")

# Visualizar a correlação entre as variáveis
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Peso', y='Comprimento', data=df)
plt.title('Correlação entre Peso e Comprimento dos Cães')
plt.xlabel('Peso (Kg)')
plt.ylabel('Comprimento (cm)')
plt.show()

```

```python
# Calcular a equação de regressão linear pelo método dos mínimos quadrados
X = df['Peso']
y = df['Comprimento']
X = sm.add_constant(X)  # Adicionar uma constante ao modelo

```

```python
model = sm.OLS(y, X).fit()
print(model.summary())

```

```python
# Obter os coeficientes da equação de regressão
intercept = model.params[0]
slope = model.params[1]
print(f"Equação de Regressão Linear: Comprimento = {intercept:.2f} + {slope:.2f} * Peso")

```

```python
# Visualizar a linha de regressão
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Peso', y='Comprimento', data=df, label='Dados Reais')
plt.plot(df['Peso'], model.predict(X), color='red', label='Linha de Regressão')
plt.title('Regressão Linear entre Peso e Comprimento dos Cães')
plt.xlabel('Peso (Kg)')
plt.ylabel('Comprimento (cm)')
plt.legend()
plt.show()
```

## `Análise e Conclusão``

`Coeficiente de Correlação de Pearson:`

1 indica uma correlação linear positiva perfeita,
-1 indica uma correlação linear negativa perfeita,
0 indica que não há correlação linear.

`Resultado`

O coeficiente de correlação de Pearson calculado entre o peso e o comprimento dos cães é exibido junto com o valor p, que indica a significância estatística da correlação.

Equação de Regressão Linear
A equação de regressão linear é calculada pelo método dos mínimos quadrados. A equação tem a forma: 

### $\text{Comprimento}$ = $\text{intercepto}$ + $\text{inclinação}$ $\times \text{Peso}$

`Visualização`

A visualização mostra a relação entre o peso e o comprimento dos cães em um gráfico de dispersão, juntamente com a linha de regressão ajustada, permitindo uma análise visual da correlação e da regressão.

`Conclusão``

Os resultados indicam a força e a direção da relação linear entre o peso e o comprimento dos cães. A equação de regressão linear fornece uma forma de prever o comprimento dos cães com base no peso. A visualização confirma que a linha de regressão ajustada se alinha bem com os dados reais, indicando uma boa adequação do modelo.

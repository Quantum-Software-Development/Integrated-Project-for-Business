```python
# importa biblioteca pandas
import pandas as pd

# importa biblioteca scipy
import scipy.stats as stats

# local do arquivo
Caminho='C:/_Academico/PUC/202401/PROJETO INTEGRADO NEGÃ“CIOS/Biscobis.xlsx'
```

```python
#Traz o arquivo para o Python
df=pd.read_excel(Caminho)
```

```python
#Gera tabela de correlaÃ§Ãµes
df.corr( )
```

```python
#Cria variÃ¡vel Dummy

dummy_x8 = pd.get_dummies(df['x8'], prefix='x8', drop_first=True)

print(dummy_x8)
```

```python
# Concatenando as variÃ¡veis dummy com o DataFrame original
df = pd.concat([df, dummy_x8], axis=1)
```

```python
#Modelo de RegressÃ£o
from statsmodels.formula.api import ols
modelo=ols('x9~x1+x2+x3+x4+x5+x6+x7+x8_2+x8_3',data=df).fit()

# Resumo do modelo
print(modelo.summary())
```

```python
#Stepwise
from stepwise_regression import step_reg
#cria dois conjuntos, o de variÃ¡veis independentes (x) e o outro com a variÃ¡vel resposta (y)
x=df[['x1','x2','x3','x4','x5','x6','x7','x8_2','x8_3']]
y=df['x9']
```

```python
#Seleciona as variÃ¡veis
backselect = step_reg.backward_regression(x, y, 0.05)
backselect
```

```python
#Modelo de RegressÃ£o
from statsmodels.formula.api import ols
modelo=ols('x9~x3+x5+x6+x7+x8_2+x8_3',data=df).fit()

# Resumo do modelo
print(modelo.summary())
```

```python

# VerificaÃ§Ã£o de DistribuiÃ§Ã£o Normal dos ResÃ­duos
# Teste de Shapiro-Wilk para normalidade dos resÃ­duos

from scipy.stats import shapiro

residuos=modelo.resid
stat, p_valor = shapiro(residuos)
print('Teste de Shapiro-Wilk para normalidade dos resÃ­duos:')
print('EstatÃ­stica de teste:', stat)
print('Valor-p:', p_valor)
```
